```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
sdg5 = pd.read_csv("/content/ods_5.csv")
sdg13 = pd.read_csv("/content/ods_13.csv")
sdg14 = pd.read_csv("/content/ods_14.csv")
sdg15 = pd.read_csv("/content/ods_15.csv")
```

Add a node for each author's id and give them it's respective name

```python
sdg5_graph = nx.Graph()
for index, row in sdg5.iterrows():
  names = row["Authors"].split(";")
  ids = row["Author(s) ID"].split(";")

  for name, id in zip(names, ids):
    sdg5_graph.add_node(id, name = name)

  for i in range(len(names)):
    for j in range(i+1, len(names)):
      sdg5_graph.add_edge(ids[i], ids[j])
```

Extract the degree of each node and put them in a list by order of appearance

```python
sdg5_deg = [j for i, j in sdg5_graph.degree()]
```

Extract the average degree of each node's neighbor and put them in a list by order of appearance

```python
sdg5_avr_deg_val = nx.average_neighbor_degree(sdg5_graph)
sdg5_avr_deg = [sdg5_avr_deg_val[i] for i in sdg5_avr_deg_val]
```

Plot a chart where the y axis is the average neighbor's degree and the x axis is the node degree

```python
plt.title("SDG 5")
plt.xlabel("Node degree")
plt.ylabel("Average neighbor degree")

plt.scatter(sdg5_deg, sdg5_avr_deg)
plt.grid()
plt.savefig("sdg5.png")
plt.show()
```
